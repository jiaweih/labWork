```python
def figure_historical_close_mean_monthly(plot_dir,basins,models,scenarios,variables,ymin_limit,ymax_limit,start,end,fig_size,fmts,
                               dict_text,dict_ymax,linewidth,legend_loc,legend_size,xlabel,label_size,tick_size,text_size,store_dir0):
    months=['Oct','Nov','Dec','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep']
    for basin in basins:
        max_list = []
        fig = plt.figure(figsize=fig_size)
        ax = fig.add_subplot(111)
        for idx,model in enumerate(models):        
            file_fldmean = plot_dir.format(basin,model)
            df = get_data_ts(file_fldmean,start,end)
            fmt = fmts[idx]
            fig_txt = "{}".format(model)
            df.groupby(lambda x: water_month(x.month)).mean().plot(style=fmt,label=fig_txt,linewidth=4)
            max_list.append(df.groupby(lambda x: water_month(x.month)).mean().max())
        y_unit = '[mm/day]'
        #ymax_limit = np.ceil(1.6*np.max(max_list))
        ymax_limit = dict_ymax[basin]
        ax.legend(loc=legend_loc,prop={'size':legend_size},**{'frameon':False})
        ax.set_xlabel(xlabel,fontsize=label_size)
        ax.set_ylabel(y_unit,fontsize=label_size)
        ax.set_ylim([ymin_limit,ymax_limit])
        ax.set_xticks(np.arange(12))
        ax.set_xlim([-0.5,11.5])
        ax.set_xticklabels(months)
        ax.tick_params(axis='both',which='major',labelsize=tick_size)
        ax.text(0.93,0.97,dict_text[basin],verticalalignment='top',horizontalalignment='right',transform=ax.transAxes,
                color='black',fontsize=text_size)
        store_dir=store_dir0.format(basin)
        fig.suptitle('Historical Precipitation',fontsize=25)
        fig.savefig(store_dir,dpi=300)
```

```python
### plot historical,rcp45,rcp85 together
def figure_monthly_mean_basin_panel(plot_dir,hist_dir,models,basins,scenarios,variables,nrow,ncol,height_ratio,wspace,hspace,
                                    ymin_limit,ymax_limit,start,end,fig_size,fmts,dic_txt,dic_var,dict_scenario,dict_ymax,dict_model,moothing_step,linewidth,
                                    linewidth_moving,legend_loc,legend_size,xlabel,label_size,tick_size,text_size,store_dir0):  
    for basin in basins:
        fig = plt.figure(figsize=fig_size)
        grid = gridspec.GridSpec(nrow,ncol,height_ratios=height_ratio,wspace=wspace,hspace=hspace)
        for idx,model in enumerate(models):
            #print grid[idx]
            #ax = fig.add_subplot(idx+1,1,idx+1)
            ax = fig.add_subplot(grid[idx])
            idy = 0
            for var in variables:
                for scenario in scenarios:
                    file_fldmean = plot_dir.format(basin,model,scenario,var)
                    #print file_fldmean
                    df = get_data_ts(file_fldmean,start,end)
                    fmt = fmts[idy]
                    fig_txt = '{} ({})'.format(dict_var[var],dict_scenario[scenario])
                    pds.rolling_mean(df,smoothing_step,center=True).plot(style=fmt,label=fig_txt,linewidth=linewidth_moving)
                    idy+=1
                file_hist = hist_dir.format(basin,model,var)
                df = get_data_ts(file_hist,start,end)
                fmt = fmts[idy]
                fig_txt = '{} ({})'.format(dict_var[var],dict_scenario['historical'])
                pds.rolling_mean(df,smoothing_step,center=True).plot(style=fmt,label=fig_txt,linewidth=3.5)
                idy+=1
            ax.legend(loc=legend_loc,prop={'size':legend_size},**{'frameon':False})
            ax.set_xlabel(xlabel,fontsize=label_size)
            y_unit = '[mm/day]'
            ax.set_ylabel(y_unit,fontsize=label_size)
            ymax_limit = dict_ymax[basin]
            ax.set_ylim([ymin_limit,ymax_limit])
            ax.tick_params(axis='both',which='major',labelsize=tick_size)
            ax.text(0.93,0.97,dict_basin[basin],verticalalignment='top',horizontalalignment='right',transform=ax.transAxes,
                    color='black',fontsize=text_size)
            
            ax.set_title(dict_model[model],fontsize=20)
       # fig.suptitle('60 Months Moving Average',fontsize = 23)
        fig.text(0.5,0.94,'60 Months Moving Average',verticalalignment='top',horizontalalignment='center',transform=ax.transAxes,
                    color='black',fontsize=text_size)
        store_dir = store_dir0.format(basin)
        fig.savefig(store_dir,dpi=300)
```

```python
def figure_mean_monthly_basin(plot_dir,basins,models,scenarios,variables,ymin_limit,ymax_limit,start,end,fig_size,fmts,
                               dict_basin,dict_var,dict_scenario,dict_ymax,dict_model,linewidth,legend_loc,legend_size,xlabel,label_size,
                               tick_size,text_size,store_dir0):
    import numpy as np
    months=['Oct','Nov','Dec','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep']
    for basin in basins:
        for model in models:
            fig = plt.figure(figsize=fig_size)
            ax = fig.add_subplot(111)
            idx = 0
            max_list = []
            for scenario in scenarios:
                for var in variables:
                    file_fldmean = plot_dir.format(basin,model,scenario,var)
                  #  print file_fldmean
                    df = get_data_ts(file_fldmean,start,end)
                    fmt=fmts[idx]
                    fig_txt = '{} ({})'.format(dict_var[var],dict_scenario[scenario])
                    df.groupby(lambda x: water_month(x.month)).mean().plot(style=fmt,label=fig_txt,linewidth=4)
                    max_list.append(df.groupby(lambda x: water_month(x.month)).mean().max())
                    idx+=1
            y_unit='[mm/day]'
           # ymax_limit = np.floor(1.6*np.max(max_list))
            ymax_limit = dict_ymax[basin]
            ax.legend(loc=2,prop={'size':legend_size},**{'frameon':False})
            ax.set_xlabel(xlabel,fontsize=label_size)
            ax.set_ylabel(y_unit,fontsize=label_size)
            ax.set_ylim([ymin_limit,ymax_limit])
            ax.set_xticks(np.arange(12))
            ax.set_xlim([-0.5,11.5])
            ax.set_xticklabels(months)
            ax.tick_params(axis='both',which='major',labelsize=tick_size)
            ax.text(0.92,0.97,dict_basin[basin],verticalalignment='top',horizontalalignment='right',transform=ax.transAxes,
                    color='black',fontsize=text_size)
            ax.set_title(dict_model[model],fontsize = 23)
            store_dir=store_dir0.format(basin,model)
            fig.savefig(store_dir,dpi=300)                       
```
