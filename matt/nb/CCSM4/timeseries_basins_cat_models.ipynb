```python
def figure_historical_close_monthly_mean(plot_dir,basins,models,scenarios,variables,ymin_limit,ymax_limit,start,end,fig_size,fmts,
                               linewidth,legend_loc,legend_size,xlabel,label_size,tick_size,text_size,store_dir0):
    for basin in basins:
        max_list = []
        fig = plt.figure(figsize=fig_size)
        ax = fig.add_subplot(111)
        for idx,model in enumerate(models):        
            file_fldmean = plot_dir.format(basin,model)
            df = get_data_ts(file_fldmean,start,end)
            fmt = fmts[idx]
            fig_txt = "{}".format(model)
            pds.rolling_mean(df,smoothing_step,center=True).plot(style=fmt,label=model,
                                                                     linewidth=linewidth_moving)
            max_list.append(df.groupby(lambda x: water_month(x.month)).mean().max())
        y_unit = '[mm/day]'
        ymax_limit = np.ceil(1.4*np.max(max_list))
        ax.legend(loc=legend_loc,prop={'size':legend_size},**{'frameon':False})
        ax.set_xlabel(xlabel,fontsize=label_size)
        ax.set_ylabel(y_unit,fontsize=label_size)
        ax.set_ylim([ymin_limit,ymax_limit])
        ax.tick_params(axis='both',which='major',labelsize=tick_size)
        ax.text(0.08,0.97,basin ,verticalalignment='top',horizontalalignment='left',transform=ax.transAxes,
                color='black',fontsize=text_size)
        store_dir=store_dir0.format(basin)
        fig.savefig(store_dir,dpi=300)
```

```python
def figure_mean_monthly_basin(plot_dir,basins,models,scenarios,variables,ymin_limit,ymax_limit,start,end,fig_size,fmts,
                               linewidth,legend_loc,legend_size,xlabel,label_size,tick_size,text_size,store_dir0):
    import numpy as np
    months=['Oct','Nov','Dec','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep']
    for basin in basins:
        for model in models:
            fig = plt.figure(figsize=fig_size)
            ax = fig.add_subplot(111)
            idx = 0
            max_list = []
            for scenario in scenarios:
                for var in variables:
                    file_fldmean = plot_dir.format(basin,model,scenario,var)
               #     print file_fldmean
                    df = get_data_ts(file_fldmean,start,end)
                    fmt=fmts[idx]
                    fig_txt = "{} ({})".format(var,scenario)
                    df.groupby(lambda x: water_month(x.month)).mean().plot(style=fmt,label=fig_txt,linewidth=4)
                    max_list.append(df.groupby(lambda x: water_month(x.month)).mean().max())
                    idx+=1
            y_unit='mm/day'
            ymax_limit = np.ceil(1.6*np.max(max_list))
            ax.legend(loc=legend_loc,prop={'size':legend_size},**{'frameon':False})
            ax.set_xlabel(xlabel,fontsize=label_size)
            ax.set_ylabel(y_unit,fontsize=label_size)
            ax.set_ylim([ymin_limit,ymax_limit])
            ax.set_xticks(np.arange(12))
            ax.set_xlim([-0.5,11.5])
            ax.set_xticklabels(months)
            ax.tick_params(axis='both',which='major',labelsize=tick_size)
            ax.text(0.08,0.97,basin,verticalalignment='top',horizontalalignment='left',transform=ax.transAxes,
                    color='black',fontsize=text_size)
            store_dir=store_dir0.format(basin,model)
            fig.savefig(store_dir,dpi=300)                       
```

```python
def figure_monthly_mean_basin_QPE_cat(plot_dir,basins,scenarios,variables,ymin_limit,ymax_limit,start,end,fig_size,
                              fmts,smoothing_step,linewidth,linewidth_moving,linewidth_axv,legend_loc,legend_size,
                              xlabel,label_size,tick_size,text_size,store_dir0):
    for basin in basins:
        for scenario in scenarios:
            fig = plt.figure(figsize=fig_size)
            ax = fig.add_subplot(111)
            idx = 0
            for model in models: 
                for var in variables:
                    file_fldmean = plot_dir.format(basin,model,scenario,var)
                    df = get_data_ts(file_fldmean,start,end)
                    fmt_smoothing = fmts[idx]
                    fig_txt_smoothing = '{} ({})'.format(var,model)                                             
                    pds.rolling_mean(df,smoothing_step,center=True).plot(style=fmt_smoothing,label=fig_txt_smoothing,
                                                                     linewidth=linewidth_moving)
                    idx+=1
            y_unit=get_yunit(var)
            div=dt.datetime(2005,12,31)
            ax.axvline(x=div,linewidth=linewidth_axv,color=color_axv)
            ax.legend(loc=legend_loc,prop={'size':legend_size},**{'frameon':False})
            ax.set_xlabel(xlabel,fontsize=label_size)
            y_unit = '[mm/day]'
            ax.set_ylabel(y_unit,fontsize=label_size)
            ax.set_ylim([ymin_limit,ymax_limit])
            ax.tick_params(axis='both',which='major',labelsize=tick_size)
            ax.text(0.06,0.97,basin,verticalalignment='top',horizontalalignment='left',transform=ax.transAxes,
                    color='black',fontsize=text_size)
            suptitle = 'Historical & {} {} months Moving Average'.format(scenario,smoothing_step)
            #fig.suptitle(suptitle,fontsize=20)
            store_dir = store_dir0.format(basin,scenario)                
            fig.savefig(store_dir,dpi=300)
```
