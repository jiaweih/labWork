```python
def figure_monthly_mean_cat_variables(plot_dir,scenarios,variables,positions,start,end,fig_size,fmt_mean,fmt_smoothing,smoothing_step,linewidth,\
                        linewidth_moving,legend_loc,legend_size,xlabel,ylabel,ymin_limit,ymax_limit,label_size,tick_size,height_ratio,width_ratio,
                        wspace,hspace,store_dir_monthly_mean0,pos_names,linewidth_axv,color_axv):
    for scenario in scenarios:
        fig = plt.figure(figsize=(fig_size))
        grid = gridspec.GridSpec(2,2,height_ratios=height_ratio,width_ratios=width_ratio,wspace=wspace,hspace=hspace)
        for idx,pos in enumerate(pos_names):
            ax = fig.add_subplot(grid[idx])
            pos_name = pos_names[idx]
            for idy,var in enumerate(variables):                  
                file_fldmean = plot_dir.format(scenario,var,scenario,pos_name)
                fig_txt_mean = pos_name + ' Monthly Mean '
                #fig_txt_smoothing = '{} Months Moving Average'.format(smoothing_step)
                fig_txt_smoothing = var
                df = get_data_ts(file_fldmean,start,end)
                pds.rolling_mean(df,smoothing_step,center=True).plot(style=fmt_smoothing[idy],label=fig_txt_smoothing,linewidth=linewidth_moving)
            div=dt.datetime(2005,12,31)          
            y_unit=get_yunit(var)
            ax.axvline(x=div,linewidth=linewidth_axv,color=color_axv)
            ax.legend(loc=legend_loc,prop={'size':legend_size})
            ax.set_xlabel(xlabel,fontsize=label_size)
            ax.set_ylabel(y_unit,fontsize=label_size)
            ax.set_ylim([ymin_limit,ymax_limit])
            ax.tick_params(axis='both',which='major',labelsize=tick_size)
            ax.text(0.08, 0.97, pos_name, verticalalignment='top', horizontalalignment='left',
                        transform=ax.transAxes, color='black', fontsize=20)
        fig.suptitle("Historical & {} {} Months Moving Average".format(scenario,smoothing_step),fontsize=25)
        store_dir = store_dir_monthly_mean0.format(scenario)
        fig.savefig(store_dir,dpi=300)
```

```python
xlabel = 'Year'
variables=['PREC','FLOW','EVAP']
parent_dir = '/raid3/jiawei/integrated_scenarios/CCSM4/simulator/{}/timeseries/division/'
plot_dir = parent_dir + '{}/fldmean_month_series_{}_hist_cat_{}'
store_dir_monthly_mean0 = '/home/ftp/pub/jiawei/Integrated_Scenarios/CCSM4/timeseries/monthly_series_quadrants_{}.png'
scenarios={'rcp45','rcp85'}
ymax_limit = 5
ymin_limit = 0
smoothing_step = 60
linewidth_axv = 3
color_axv = 'k'
ylabel = ''
fmt_smoothing = ['b','g','c']
figure_monthly_mean_cat_variables(plot_dir,scenarios,variables,positions,start,end,fig_size,fmt_mean,fmt_smoothing,smoothing_step,
                    linewidth,linewidth_moving,legend_loc,legend_size,xlabel,ylabel,ymin_limit,ymax_limit,label_size,
                    tick_size,height_ratio,width_ratio,wspace,hspace,store_dir_monthly_mean0,pos_names,linewidth_axv,color_axv)
```

```python
def figure_mean_monthly_together_variables(plot_dir,scenarios,variables,positions,ymin_limit,ymax_limit,start,end,fig_size,fmts,\
                              linewidth,legend_loc,legend_size,xlabel,label_size,tick_size,height_ratio,
                                width_ratio,wspace,hspace,text_size,pos_names,store_dir):
    grid = gridspec.GridSpec(2,2,height_ratios=height_ratio,width_ratios=width_ratio,wspace=wspace,hspace=hspace)
    fig = plt.figure(figsize=(fig_size))
    monthes=['Oct','Nov','Dec','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep']
    for idx,pos in enumerate(positions):
        ax = fig.add_subplot(grid[idx])
        pos_name = pos_names[idx]
        idy = 0
        for var in variables:       
            for scenario in scenarios:
                file_fldmean = plot_dir.format(scenario,var,pos_name)
                fmt = fmts[idy]
                fig_txt = "{} ({})".format(var,scenario)
                df = get_data_ts(file_fldmean,start,end)
                df.groupby(lambda x: water_month(x.month)).mean().plot(style=fmt,label=fig_txt,linewidth=linewidth)
                idy+=1
        y_unit = get_yunit(var)
        ylabel = get_var_name(var) 
        ax.legend(loc=legend_loc,prop={'size':legend_size})
        ax.set_xlabel(xlabel,fontsize=label_size)
        ax.set_ylabel(y_unit,fontsize=label_size)
        ax.set_ylim([ymin_limit,ymax_limit])
        ax.set_xticks(np.arange(12))
        ax.set_xlim([-0.5,11.5])
        ax.set_xticklabels(monthes)
        ax.tick_params(axis='both',which='major',labelsize=tick_size)
        ax.text(0.08, 0.97, pos_name, verticalalignment='top', horizontalalignment='left',
                    transform=ax.transAxes, color='black', fontsize=text_size)
    fig.suptitle("Mean Monthly of Precipitation and Total Runoff",fontsize=25)
    fig.savefig(store_dir,dpi=300)
```

```python
from functions_timeseries import figure_mean_monthly_together
xlabel = 'Month'
variables=['PREC','FLOW']
scenarios = ['historical','rcp45','rcp85']
ymin_limit = 0
ymax_limit = 6
fmts = ['k','b-','r-','k--','b--','r--','k-.','b-.','r-.']
text_size = 24
linewidth = 4
parent_dir = '/raid3/jiawei/integrated_scenarios/CCSM4/simulator/{}/timeseries/division/'
plot_dir = parent_dir + '{}/fldmean_month_series_{}'
store_dir = '/home/ftp/pub/jiawei/Integrated_Scenarios/CCSM4/timeseries/mean_monthly_quadrants_all.png'
figure_mean_monthly_together_variables(plot_dir,scenarios,variables,positions,ymin_limit,ymax_limit,start,end,fig_size,fmts,\
                              linewidth,legend_loc,legend_size,xlabel,label_size,tick_size,height_ratio,
                              width_ratio,wspace,hspace,text_size,pos_names,store_dir)
```
