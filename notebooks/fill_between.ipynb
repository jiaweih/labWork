```python
### plot historical,rcp45,rcp85 together
def figure_monthly_mean_basin_panel(plot_dir,hist_dir,models,basins,scenarios,variables,nrow,ncol,height_ratio,wspace,hspace,
                                    ymin_limit,ymax_limit,start,end,fig_size,fmts,dic_txt,dic_var,dict_scenario,dict_ymax,dict_model,
                                    moothing_step,linewidth,
                                    linewidth_moving,legend_loc,legend_size,xlabel,label_size,tick_size,text_size,store_dir0):  
    for basin in basins:
        fig = plt.figure(figsize=fig_size)
        grid = gridspec.GridSpec(nrow,ncol,height_ratios=height_ratio,wspace=wspace,hspace=hspace)
        for idx,model in enumerate(models):
            #print grid[idx]
            #ax = fig.add_subplot(idx+1,1,idx+1)
            ax = fig.add_subplot(grid[idx])
            idy = 0
            for var in variables:
                for scenario in scenarios:
                    file_fldmean = plot_dir.format(basin,model,scenario,var)
                    #print file_fldmean
                    df = get_data_ts(file_fldmean,start,end)
                    fmt = fmts[idy]
                    fig_txt = '{} ({})'.format(dict_var[var],dict_scenario[scenario])
                    pds.rolling_mean(df,smoothing_step,center=True).plot(style=fmt,label=fig_txt,linewidth=linewidth_moving)
                    idy+=1
                file_hist = hist_dir.format(basin,model,var)
                df = get_data_ts(file_hist,start,end)
                fmt = fmts[idy]
                fig_txt = '{} ({})'.format(dict_var[var],dict_scenario['historical'])
                pds.rolling_mean(df,smoothing_step,center=True).plot(style=fmt,label=fig_txt,linewidth=3.5)
                idy+=1
            ax.legend(loc=legend_loc,prop={'size':legend_size},**{'frameon':False})
            ax.set_xlabel(xlabel,fontsize=label_size)
            y_unit = '[mm/day]'
            ax.set_ylabel(y_unit,fontsize=label_size)
            ymax_limit = dict_ymax[basin]
            ax.set_ylim([ymin_limit,ymax_limit])
            ax.tick_params(axis='both',which='major',labelsize=tick_size)
            ax.text(0.93,0.97,dict_basin[basin],verticalalignment='top',horizontalalignment='right',transform=ax.transAxes,
                    color='black',fontsize=text_size)
            
            ax.set_title(dict_model[model],fontsize=20)
       # fig.suptitle('60 Months Moving Average',fontsize = 23)
        fig.text(0.5,0.94,'60 Months Moving Average',verticalalignment='top',horizontalalignment='center',transform=ax.transAxes,
                    color='black',fontsize=text_size)
        store_dir = store_dir0.format(basin)
        fig.savefig(store_dir,dpi=300)
```

```python
def figure_mean_monthly_basin_fill(ifile,basins,scenarios,var,stats,nrow,ncol,height_ratio,hspace,fig_size,
                                   step,line_width):
    min_stat = stats['min']
    max_stat = stats['max']
    mean_stat = stats['mean']
    for basin in basins:
        grid = gridspec.GridSpec(nrow,ncol,height_ratios=height_ratio,hspace=hspace)
        fig = plt.figure(figsize=fig_size)
        for idx,scenario in enumerate(scenarios):
            ax = fig.add_subplot(grid[idx])
            
            max_file = ifile.format(basin,max_stat,scenario,var)
            print max_file
            df_max = get_data_ts(max_file)
            max_value = pds.rolling_mean(df_max,step,center=True)
            #print max_value
            #pds.rolling_mean(df_max,step,center=True).plot()
            min_file = ifile.format(basin,min_stat,scenario,var)
            df_min = get_data_ts(min_file)
            min_value = pds.rolling_mean(df_min,step,center=True)
            
                  
            mean_file = ifile.format(basin,mean_stat,scenario,var)
            df = get_data_ts(mean_file)
            #pds.rolling_mean(df,step,center=True).plot(style=fmt,label=fig_txt,linewidth=line_width)
            pds.rolling_mean(df,step,center=True).plot(linewidth=line_width)
            ax.fill_between(min_value.index,min_value,max_value,color = 'grey') 
        
```
