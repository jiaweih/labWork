```python
def figure_mean_monthly_basin(plot_dir,basins,scenarios,variables,ymin_limit,ymax_limit,start,end,fig_size,fmts,
                               linewidth,legend_loc,legend_size,xlabel,label_size,tick_size,text_size,store_dir):
    months=['Oct','Nov','Dec','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep']
    for basin in basins:
        fig = plt.figure(figsize=fig_size)
        ax = fig.add_subplot(111)
        idx = 0
        for scenario in scenarios:
            for var in variables:
                file_fldmean = plot_dir.format(scenario,basin,var)
                df = get_data_ts(file_fldmean,start,end)
                fmt=fmts[idx]
                fig_txt = "{} ({})".format(var,scenario)
                df.groupby(lambda x: water_month(x.month)).mean().plot(style=fmt,label=fig_txt,linewidth=4)
                idx+=1
        y_unit='[mm/day]'
        ax.legend(loc=legend_loc,prop={'size':legend_size},**{'frameon':False})
        ax.set_xlabel(xlabel,fontsize=label_size)
        ax.set_ylabel(y_unit,fontsize=label_size)
        ax.set_ylim([ymin_limit,ymax_limit])
        ax.set_xticks(np.arange(12))
        ax.set_xlim([-0.5,11.5])
        ax.set_xticklabels(months)
        ax.tick_params(axis='both',which='major',labelsize=tick_size)
        ax.text(0.08,0.97,basin,verticalalignment='top',horizontalalignment='left',transform=ax.transAxes,
                color='black',fontsize=text_size)
        fig.savefig(store_dir,dpi=300)                       
```

```python
xlabel = 'Month'
variables=['PREC','FLOW']
scenarios = ['historical','rcp45','rcp85']
basins=['Dalles']
fig_size = (15,10)
ymin_limit = 0
ymax_limit = 5
fmts = ['k-','b-','r-','k--','b--','r--','k-.','b-.','r-.']
text_size = 24
plot_dir = parent_dir + '/{}/fldmean_month_series_{}'
store_dir = '/home/ftp/pub/jiawei/Integrated_Scenarios/CCSM4/timeseries/basins/dalle/mean_monthly_dalle.png'
figure_mean_monthly_basin(plot_dir,basins,scenarios,variables,ymin_limit,ymax_limit,start,end,fig_size,fmts,
                               linewidth,legend_loc,legend_size,xlabel,label_size,tick_size,text_size,store_dir)
```

<center>Monthly Mean of the Dalles for Historical</center>

```python
def figure_monthly_mean_basin(plot_dir,basins,scenarios,variables,ymin_limit,ymax_limit,start,end,fig_size,fmt_mean,
                              fmt_smoothing,smoothing_step,linewidth,linewidth_moving,legend_loc,legend_size,
                              xlabel,label_size,tick_size,text_size,store_dir):
    for basin in basins:
        for scenario in scenarios:
            for var in variables:
                fig = plt.figure(figsize=fig_size)
                ax = fig.add_subplot(111)
                file_fldmean = plot_dir.format(scenario,basin,var)
                df = get_data_ts(file_fldmean,start,end)
                fig_txt_mean = ' Monthly Mean ' + '(' + scenario + ')'
                fig_txt_smoothing = '{} Months Moving Average'.format(smoothing_step)
                df.plot(style=fmt_mean,label=fig_txt_mean)                                              
                pds.rolling_mean(df,smoothing_step,center=True).plot(style=fmt_smoothing,label=fig_txt_smoothing,
                                                                     linewidth=linewidth_moving)
                ylabel = get_var_name(var)
                y_unit=get_yunit(var)
                ax.legend(loc=legend_loc,prop={'size':legend_size})
                ax.set_xlabel(xlabel,fontsize=label_size)
                ax.set_ylabel(ylabel+y_unit,fontsize=label_size)
                ax.set_ylim([ymin_limit,ymax_limit])
                ax.tick_params(axis='both',which='major',labelsize=tick_size)
                ax.text(0.08,0.97,basin,verticalalignment='top',horizontalalignment='left',transform=ax.transAxes,
                        color='black',fontsize=text_size)
                store_dir = store_dir.format(basin,var)                
                fig.savefig(store_dir,dpi=300)
```

```python
def figure_annual_mean_basin(plot_dir,basins,scenarios,variables,fig_size,fmt_mean,fmt_smoothing,smoothing_step,
                             linewidth,linewidth_moving,legend_loc,legend_size,xlabel,ymin_limit,ymax_limit,label_size,
                             tick_size,store_dir):
    for basin in basins:
        for scenario in scenarios:
            for var in variables:
                fig = plt.figure(figsize=fig_size)
                ax = fig.add_subplot(111)
                file_fldmean = plot_dir.format(scenario,basin,var)
                df = get_data_ts(file_fldmean,start,end)
                fig_txt_mean = ' Annual Mean ' + '(' + scenario + ')'
                fig_txt_smoothing = '{} Years Moving Average'.format(smoothing_step)
                df.resample('A',how='mean').plot(style=fmt_mean,label=fig_txt_mean)
                df_mean = df.resample('A',how='mean')
                pds.rolling_mean(df_mean,smoothing_step,center=True).plot(style=fmt_smoothing,\
                                                                          label=fig_txt_smoothing,linewidth=linewidth_moving)
                ax.legend(loc=legend_loc,prop={'size':legend_size})
                ax.set_xlabel(xlabel,fontsize=label_size)
                ylabel = get_var_name(var)
                y_unit=get_yunit(var)
                ax.set_ylabel(ylabel+y_unit,fontsize=label_size)
                ax.set_ylim([ymin_limit,ymax_limit])
                ax.tick_params(axis='both',which='major',labelsize=tick_size)
                ax.text(0.08,0.97,basin,verticalalignment='top',horizontalalignment='left',transform=ax.transAxes,
                        color='black',fontsize=text_size)
                store_dir = store_dir.format(basin,var)
                fig.savefig(store_dir,dpi=300)
```

```python
def figure_monthly_mean_cat_variables(plot_dir,scenarios,variables,positions,start,end,fig_size,fmt_mean,fmt_smoothing,smoothing_step,linewidth,\
                        linewidth_moving,legend_loc,legend_size,xlabel,ylabel,ymin_limit,ymax_limit,label_size,tick_size,height_ratio,width_ratio,
                        wspace,hspace,store_dir_monthly_mean0,pos_names,linewidth_axv,color_axv):
    for scenario in scenarios:
        fig = plt.figure(figsize=(fig_size))
        grid = gridspec.GridSpec(2,2,height_ratios=height_ratio,width_ratios=width_ratio,wspace=wspace,hspace=hspace)
        for idx,pos in enumerate(pos_names):
            ax = fig.add_subplot(grid[idx])
            pos_name = pos_names[idx]
            for idy,var in enumerate(variables):                  
                file_fldmean = plot_dir.format(scenario,var,scenario,pos_name)
                fig_txt_mean = pos_name + ' Monthly Mean '
                #fig_txt_smoothing = '{} Months Moving Average'.format(smoothing_step)
                fig_txt_smoothing = var
                df = get_data_ts(file_fldmean,start,end)
                pds.rolling_mean(df,smoothing_step,center=True).plot(style=fmt_smoothing[idy],label=fig_txt_smoothing,linewidth=linewidth_moving)
            div=dt.datetime(2005,12,31)          
            y_unit=get_yunit(var)
            ax.axvline(x=div,linewidth=linewidth_axv,color=color_axv)
            ax.legend(loc=legend_loc,prop={'size':legend_size})
            ax.set_xlabel(xlabel,fontsize=label_size)
            ax.set_ylabel(y_unit,fontsize=label_size)
            ax.set_ylim([ymin_limit,ymax_limit])
            ax.tick_params(axis='both',which='major',labelsize=tick_size)
            ax.text(0.08, 0.97, pos_name, verticalalignment='top', horizontalalignment='left',
                        transform=ax.transAxes, color='black', fontsize=20)
        fig.suptitle("Historical & {} {} Months Moving Average".format(scenario,smoothing_step),fontsize=25)
        store_dir = store_dir_monthly_mean0.format(scenario)
        fig.savefig(store_dir,dpi=300)
```
