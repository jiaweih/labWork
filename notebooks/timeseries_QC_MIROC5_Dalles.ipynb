```python
def figure_monthly_mean_basin(plot_dir,basins,scenarios,variables,ymin_limit,ymax_limit,start,end,fig_size,fmt_mean,
                              fmt_smoothing,smoothing_step,linewidth,linewidth_moving,legend_loc,legend_size,
                              xlabel,label_size,tick_size,text_size,store_dir):
    for basin in basins:
        for scenario in scenarios:
            for var in variables:
                fig = plt.figure(figsize=fig_size)
                ax = fig.add_subplot(111)
                file_fldmean = plot_dir.format(scenario,basin,var)
                df = get_data_ts(file_fldmean,start,end)
                fig_txt_mean = ' Monthly Mean ' + '(' + scenario + ')'
                fig_txt_smoothing = '{} Months Moving Average'.format(smoothing_step)
                df.plot(style=fmt_mean,label=fig_txt_mean)                                              
                pds.rolling_mean(df,smoothing_step,center=True).plot(style=fmt_smoothing,label=fig_txt_smoothing,
                                                                     linewidth=linewidth_moving)
                ylabel = get_var_name(var)
                y_unit=get_yunit(var)
                ax.legend(loc=legend_loc,prop={'size':legend_size})
                ax.set_xlabel(xlabel,fontsize=label_size)
                ax.set_ylabel(ylabel+y_unit,fontsize=label_size)
                ax.set_ylim([ymin_limit,ymax_limit])
                ax.tick_params(axis='both',which='major',labelsize=tick_size)
                ax.text(0.08,0.97,basin,verticalalignment='top',horizontalalignment='left',transform=ax.transAxes,
                        color='black',fontsize=text_size)
                store_dir = store_dir.format(basin,var)                
                fig.savefig(store_dir,dpi=300)
```

```python
def figure_monthly_mean_basin_QPE(plot_dir,basins,scenarios,variables,ymin_limit,ymax_limit,start,end,fig_size,
                              fmts,smoothing_step,linewidth,linewidth_moving,legend_loc,legend_size,
                              xlabel,label_size,tick_size,text_size,store_dir):
    for basin in basins:
        for scenario in scenarios:
            fig = plt.figure(figsize=fig_size)
            ax = fig.add_subplot(111)
            for idx,var in enumerate(variables):             
                file_fldmean = plot_dir.format(scenario,basin,var)
                df = get_data_ts(file_fldmean,start,end)
                fmt_smoothing = fmts[idx]
                fig_txt_smoothing = '{}'.format(var)                                             
                pds.rolling_mean(df,smoothing_step,center=True).plot(style=fmt_smoothing,label=fig_txt_smoothing,
                                                                     linewidth=linewidth_moving)
            y_unit=get_yunit(var)
            ax.legend(loc=legend_loc,prop={'size':legend_size})
            ax.set_xlabel(xlabel,fontsize=label_size)
            ax.set_ylabel(y_unit,fontsize=label_size)
            ax.set_ylim([ymin_limit,ymax_limit])
            ax.tick_params(axis='both',which='major',labelsize=tick_size)
            ax.text(0.06,0.97,basin,verticalalignment='top',horizontalalignment='left',transform=ax.transAxes,
                    color='black',fontsize=text_size)
            suptitle = '60 Months Moving Average of Precipitation, Evaporation and Flow for Historical'
            fig.suptitle(suptitle,fontsize=20)
            store_dir = store_dir.format(basin,var)                
            fig.savefig(store_dir,dpi=300)
```

```python
def figure_annual_mean_basin(plot_dir,basins,scenarios,variables,fig_size,fmt_mean,fmt_smoothing,smoothing_step,
                             linewidth,linewidth_moving,legend_loc,legend_size,xlabel,ymin_limit,ymax_limit,label_size,
                             tick_size,store_dir):
    for basin in basins:
        for scenario in scenarios:
            for var in variables:
                fig = plt.figure(figsize=fig_size)
                ax = fig.add_subplot(111)
                file_fldmean = plot_dir.format(scenario,basin,var)
                df = get_data_ts(file_fldmean,start,end)
                fig_txt_mean = ' Annual Mean ' + '(' + scenario + ')'
                fig_txt_smoothing = '{} Years Moving Average'.format(smoothing_step)
                df.resample('A',how='mean').plot(style=fmt_mean,label=fig_txt_mean)
                df_mean = df.resample('A',how='mean')
                pds.rolling_mean(df_mean,smoothing_step,center=True).plot(style=fmt_smoothing,\
                                                                          label=fig_txt_smoothing,linewidth=linewidth_moving)
                ax.legend(loc=legend_loc,prop={'size':legend_size})
                ax.set_xlabel(xlabel,fontsize=label_size)
                ylabel = get_var_name(var)
                y_unit=get_yunit(var)
                ax.set_ylabel(ylabel+y_unit,fontsize=label_size)
                ax.set_ylim([ymin_limit,ymax_limit])
                ax.tick_params(axis='both',which='major',labelsize=tick_size)
                ax.text(0.08,0.97,basin,verticalalignment='top',horizontalalignment='left',transform=ax.transAxes,
                        color='black',fontsize=text_size)
                store_dir = store_dir.format(basin,var)
                fig.savefig(store_dir,dpi=300)
```
